## Line plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var_1 = [1,2,3,4,5,6,7]
var_2 = [2,3,4,1,5,6,7]

data=pd.DataFrame({
    "var_1":var_1,
    "var_2":var_2
})

sns.lineplot(data=data,x="var_1",y="var_2")

plt.show()  # For removing the x_label and y_label shown on the top side of the diagram 
```

```python
# How the matplotlib show the output
plt.plot(var_1,var_2)
plt.show()
```

```python
sns.lineplot(data=data_1,x="who",y="alive")
plt.show()
```

```python
# hue represent with respect to basically 
sns.lineplot(data=data_1,x="who",y="alive",hue="class")
plt.show()
```

```python
# Size parameter : we can customize the size
sns.lineplot(data=data_1,x="who",y="alive",hue="class",size=60)
```

```python
# style parameter : change the style of any label
sns.lineplot(data=data_1,x="who",y="alive",hue="class",style="class")
```
